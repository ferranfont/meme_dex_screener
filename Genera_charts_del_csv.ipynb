```python
import glob
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import os

def read_file_with_fallback_encodings(file_path, encodings=['utf-8', 'latin1', 'ISO-8859-1']):
    for encoding in encodings:
        try:
            with open(file_path, 'r', encoding=encoding) as f:
                print(f"Leyendo {file_path} con codificación {encoding}")
                return f.readlines()
        except UnicodeDecodeError as e:
            print(f"Error de Unicode con codificación {encoding} en {file_path}: {e}")
        except Exception as e:
            print(f"Error inesperado al leer {file_path} con codificación {encoding}: {e}")
            raise  # Re-lanza la excepción para detener el script
    raise UnicodeDecodeError(f"No se pudo leer el archivo {file_path} con las codificaciones proporcionadas.")

def plot_charts_from_csv_v6(input_folder, output_folder):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)
    print(f"Buscando archivos en {input_folder}")
    csv_files = glob.glob(os.path.join(input_folder, '*.csv'))
    
    for file_path in csv_files:
        print(f"Procesando {file_path}")
        try:
            lines = read_file_with_fallback_encodings(file_path)
            first_line = lines[0].strip()
            url_comment = first_line.replace("# ", "")
            
            from io import StringIO
            df_content = StringIO("".join(lines[2:]))
            df = pd.read_csv(df_content, names=['time', 'price', 'volume', 'liquidity', 'fdv', 'buys', 'sells', 'age'])
            df['time'] = pd.to_datetime(df['time']).dt.strftime('%d %H:%M')
            
            base_filename = os.path.basename(file_path).replace('.csv', '')
            
            fig = plt.figure(figsize=(10, 6))
            gs = gridspec.GridSpec(2, 1, height_ratios=[3, 1])
            ax1 = plt.subplot(gs[0])
            ax2 = plt.subplot(gs[1], sharex=ax1)
            
            ax1_right = ax1.twinx()
            ax1_right.plot(df['time'], df['price'], label='Price', color='blue')
            ax1_right.set_ylabel('Price', color='blue')
            
            ax1.plot(df['time'], df['liquidity'], label='Liquidity', color='tab:orange')
            ax1.set_ylabel('Liquidity', color='tab:orange')
            
            ax2.plot(df['time'], df['volume'], label='Volume', color='tab:green')
            ax2.set_xlabel('Time')
            ax2.set_ylabel('Volume', color='tab:green')
            
            plt.xticks(rotation=45)
            plt.suptitle(f'{base_filename} - {url_comment}', fontsize=8)
            plt.tight_layout(rect=[0, 0.03, 1, 0.95])
            
            output_path = os.path.join(output_folder, f"{base_filename}.png")
            plt.savefig(output_path)
            plt.close()
            print(f"Archivo {output_path} generado con éxito.")
        except Exception as e:
            print(f"Error procesando {file_path}: {e}")

# Asegúrate de reemplazar 'input' y 'charts' con las rutas correctas a tus carpetas de entrada y salida
plot_charts_from_csv_v6('input', 'charts')

```
